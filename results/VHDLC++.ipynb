```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from matplotlib.gridspec import GridSpec

# ── 1. PSEI ACTUAL 2022 ──────────────────────────────────────────────────────
psei = pd.read_csv(r"C:\Users\rakkh\OneDrive\Desktop\RP\PSEi Composite Historical Data (1).csv")
psei['Date'] = pd.to_datetime(psei['Date'], format='%m/%d/%Y')
psei2022 = psei[psei['Date'].dt.year == 2022].sort_values('Date').copy()
psei2022['Actual'] = psei2022['Price'].str.replace(',', '').astype(float)
psei2022 = psei2022[['Date', 'Actual']].reset_index(drop=True)

# ── 2. C++ FORECAST ───────────────────────────────────────────────────────────
fc = pd.read_csv(r"C:\Users\rakkh\OneDrive\Desktop\RP\.vscode\forecast_2017_2021_vs_2022_add_vs_mul_log_vs_orig_GA.csv", nrows=12)
cpp_forecast = fc['Forecast_Orig'].values

# ── 3. VHDL ILA DATA ─────────────────────────────────────────────────────────
ila = pd.read_csv(r"C:\Users\rakkh\hw_ilaidea\iladata.csv")
ila_subset = ila.iloc[401:413]
vhdl_forecast = pd.to_numeric(ila_subset['forecast_out_s[31:0]'], errors='coerce').values / 65536

# ── 4. COMBINE ────────────────────────────────────────────────────────────────
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df = pd.DataFrame({
    'Month':         months,
    'Actual':        psei2022['Actual'].values,
    'CPP_Forecast':  cpp_forecast,
    'VHDL_Forecast': vhdl_forecast
})

# ── 5. METRICS ────────────────────────────────────────────────────────────────
df['Error_CPP']   = df['Actual'] - df['CPP_Forecast']
df['Error_VHDL']  = df['Actual'] - df['VHDL_Forecast']
df['AbsDev']      = (df['CPP_Forecast'] - df['VHDL_Forecast']).abs()

mae_cpp   = df['Error_CPP'].abs().mean()
mae_vhdl  = df['Error_VHDL'].abs().mean()
rmse_cpp  = np.sqrt((df['Error_CPP']**2).mean())
rmse_vhdl = np.sqrt((df['Error_VHDL']**2).mean())
max_dev   = df['AbsDev'].max()

print(df[['Month','Actual','CPP_Forecast','VHDL_Forecast']].to_string(index=False))
print(f'\nC++  → MAE: {mae_cpp:.3f}  |  RMSE: {rmse_cpp:.3f}')
print(f'VHDL → MAE: {mae_vhdl:.3f}  |  RMSE: {rmse_vhdl:.3f}')
print(f'Max C++ vs VHDL deviation: {max_dev:.4f} pts')

# ── 6. PLOT ───────────────────────────────────────────────────────────────────
fig = plt.figure(figsize=(13, 9))
fig.patch.set_facecolor('white')

gs = GridSpec(2, 1, height_ratios=[3, 1], hspace=0.35)

x = np.arange(len(months))

# ── TOP PANEL ─────────────────────────────────────────────────────────────────
ax1 = fig.add_subplot(gs[0])
ax1.set_facecolor('#f8f9fa')

ax1.plot(x, df['Actual'],
         color='#1f77b4', marker='o', linewidth=2.2,
         markersize=6, label='Actual PSEI', zorder=3)

ax1.plot(x, df['CPP_Forecast'],
         color='#d62728', marker='s', linewidth=2,
         markersize=6, linestyle='--', label='C++ Forecast', zorder=4)

ax1.plot(x, df['VHDL_Forecast'],
         color='#2ca02c', marker='^', linewidth=2,
         markersize=6, linestyle=':', label='VHDL Forecast', zorder=5)

ax1.set_title('C++ vs VHDL Holt-Winters Forecast Comparison (2022)',
              fontsize=14, fontweight='bold', pad=14)
ax1.set_ylabel('PSEI Index Points', fontsize=11)
ax1.set_xticks(x)
ax1.set_xticklabels(months, fontsize=10)
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda val, _: f'{val:,.0f}'))
ax1.grid(True, linestyle='--', alpha=0.5, color='gray')
ax1.legend(loc='upper right', fontsize=10, framealpha=0.9)

# ── BOTTOM PANEL — deviation bars ─────────────────────────────────────────────
ax2 = fig.add_subplot(gs[1])
ax2.set_facecolor('#f8f9fa')

bar_color = '#9999cc'
bars = ax2.bar(x, df['AbsDev'], color=bar_color, width=0.55,
               edgecolor='white', linewidth=0.5, zorder=3)

# Annotate max deviation
max_idx = df['AbsDev'].idxmax()
ax2.annotate(
    f'Max deviation = {max_dev:.4f} pts  (<0.001 pts)',
    xy=(max_idx, max_dev),
    xytext=(max_idx - 3.2, max_dev * 0.85),
    fontsize=9,
    bbox=dict(boxstyle='round,pad=0.3', facecolor='white',
              edgecolor='#888888', linewidth=0.8),
    arrowprops=dict(arrowstyle='->', color='#555555', lw=1.2)
)

ax2.set_ylabel('|C++ − VHDL|\n(index pts)', fontsize=10)
ax2.set_xlabel('Month (2022)', fontsize=11)
ax2.set_xticks(x)
ax2.set_xticklabels(months, fontsize=10)
ax2.yaxis.set_major_formatter(mticker.FormatStrFormatter('%.4f'))
ax2.grid(True, linestyle='--', alpha=0.4, color='gray', axis='y')
ax2.set_ylim(bottom=0)

# ── FOOTER METRICS ────────────────────────────────────────────────────────────
fig.text(0.13, 0.01,
         f'C++ Forecast:   MAE = {mae_cpp:.3f},  RMSE = {rmse_cpp:.3f}   '
         f'     VHDL Forecast:  MAE = {mae_vhdl:.3f},  RMSE = {rmse_vhdl:.3f}',
         fontsize=9.5, color='#333333',
         ha='left', va='bottom',
         bbox=dict(facecolor='#f0f0f0', edgecolor='#cccccc',
                   boxstyle='round,pad=0.4', linewidth=0.8))

# ── SAVE ──────────────────────────────────────────────────────────────────────
out = r"C:\Users\rakkh\OneDrive\Desktop\RP\vhdl_vs_cpp_forecast.png"
plt.savefig(out, dpi=150, bbox_inches='tight', facecolor='white')
print(f'\nPlot saved to: {out}')
plt.show()

# ── SAVE CSV ──────────────────────────────────────────────────────────────────
csv_out = r"C:\Users\rakkh\OneDrive\Desktop\RP\combined_forecast_data.csv"
df.to_csv(csv_out, index=False)
print(f'CSV saved to:  {csv_out}')
```
