```python
import pandas as pd
import matplotlib.pyplot as plt

# ----- Load your CSV file -----
# Change the file path if needed
df = pd.read_csv(r"C:\Users\rakkh\OneDrive\Desktop\RP\.vscode\forecast_REAL.csv")

# ----- Plot -----
plt.figure(figsize=(12, 6))


plt.plot(df["Index"], df["Actual"], label="Actual", linewidth=2)


# ----- Decorations -----
plt.title("Actual vs Forecast Values", fontsize=16)
plt.xlabel("Index", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)

# ----- Show -----
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ----- Load your CSV file -----
df = pd.read_csv(r"C:\Users\rakkh\OneDrive\Desktop\RP\.vscode\forecast_REAL.csv")

# ----- Calculate Error -----
df["Error"] = df["Actual"] - df["Forecast"]

# ----- Plot -----
plt.figure(figsize=(12, 6))

plt.plot(df["Index"], df["Actual"], label="Actual", linewidth=2)
plt.plot(df["Index"], df["Forecast"], label="Forecast", linewidth=2)
plt.plot(df["Index"], df["Error"], label="Actual - Forecast (Error)", linewidth=2)

# ----- Decorations -----
plt.title("Actual vs Forecast vs Error", fontsize=16)
plt.xlabel("Index", fontsize=14)
plt.ylabel("Value", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)

# ----- Show -----
plt.tight_layout()
plt.show()
```
